```python
import gzip
import networkx as nx

# Create an undirected graph
G = nx.Graph()

# Open the gzipped file
with gzip.open('../data/com-amazon.ungraph.txt.gz', 'rt') as f:
    for line in f:
        # Skip comments
        if line.startswith('#'):
            continue
        # Split nodes and add edge
        u, v = line.strip().split()
        G.add_edge(u, v)

print(f"Number of nodes: {G.number_of_nodes()}")
print(f"Number of edges: {G.number_of_edges()}")
```

```python
# Read top 5000 communities
communities_file = '../data/com-amazon.top5000.cmty.txt.gz'
node_to_category = {}

with gzip.open(communities_file, 'rt') as f:
    for idx, line in enumerate(f):
        nodes = line.strip().split()
        for node in nodes:
            node_to_category[node] = f"category_{idx}"

# Example: attach as node attribute
nx.set_node_attributes(G, node_to_category, name='category')
```

```python
import pickle

with open('../output/amazon_graph.pkl', 'wb') as f:
    pickle.dump(G, f)
```
